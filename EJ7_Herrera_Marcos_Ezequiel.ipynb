```python
import numpy as np
import matplotlib.pyplot as plt


def function_g(x):
    return (x-1)**8


def function_f(x):
    return x**8 - 8 * x**7 + 28 * x**6 - 56 * x**5 + 70 * x**4 - 56 * x**3 + 28 * x**2 - 8 * x + 1
    

def run():
    print(function_f(3))
    print(function_g(3))
    print('Cantidad de puntos evaluados en ambos casos = 100')

    x = np.linspace(0.999,1.001,100)
    plt.plot(x,function_g(x))
    plt.show()

    y = np.linspace(0.999,1.001,100)
    plt.plot(y,function_g(y))
    plt.show()


if __name__ == '__main__':
    run()
```
